### Ques1

Scrape the details of most viewed videos on YouTube from Wikipedia. Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: A) Rank B) Name C) Artist D) Upload date E) Views

```python
driver = webdriver.Chrome('chromedriver.exe')
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')

rank = []
video_names = []
uploader = []
views_billion = []
upload_date = []


time.sleep(6)

#Song Title
for i in range(1,31):
    video_names.append(driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]//tr')[i].find_elements_by_xpath('.//td')[1].text.split('"')[1])

#Uploader
for i in range(1,31):
    uploader.append(driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]//tr')[i].find_elements_by_xpath('.//td')[2].text)
  
#Views in Billion
for i in range(1,31):
    views_billion.append(driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]//tr')[i].find_elements_by_xpath('.//td')[3].text)

#Rank    
for i in range(1,31):
    rank.append(driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]//tr')[i].find_elements_by_xpath('.//td')[0].text)
    
#Upload Date
for i in range(1,31):
    upload_date.append(driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]//tr')[i].find_elements_by_xpath('.//td')[4].text)
    
df = pd.DataFrame({'Rank':rank, 'Video_Name':video_names, 'Artist':uploader, 'Views':views_billion, 'upload_date':upload_date})
df
```

### Ques2

Scrape the details team India’s international fixtures from bcci.tv. Url = https://www.bcci.tv/. You need to find following details: A) Match title (I.e. 1 st ODI) B) Series C) Place D) Date E) Time

```python
driver = webdriver.Chrome('chromedriver.exe')
driver.get('https://www.bcci.tv/')

#fetching button link
driver.get(driver.find_element_by_xpath('//a[@class="moreMatchesLink"]').get_attribute('href'))

time.sleep(5)
series = [series.text.split('\n')[0] for series in driver.find_elements_by_xpath('//div[@class="fixture-card-top"]')] 
date = [title.text.split('\n')[1] for title in driver.find_elements_by_xpath('//div[@class="fixture-card-top"]')] 
timee = [title.text.split('\n')[2] for title in driver.find_elements_by_xpath('//div[@class="fixture-card-top"]')] 
place = [place.text.split(' - ')[1] for place in driver.find_elements_by_xpath('//div[@class="fix-place ng-binding ng-scope"]')]
title = [title.text.split(' - ')[1] for title in driver.find_elements_by_xpath('//div[@class="fix-place ng-binding ng-scope"]')]

df = pd.DataFrame({'Match title':title, 'Series':series, 'Place':place, 'Date':date, 'Time':timee})
df
```

## Ques-4

Scrape the details of State-wise GDP of India from statisticstime.com. Url = http://statisticstimes.com/ You have to find following details: A) Rank B) State C) GSDP(18-19) D) GSDP(17-18) E) Share(2017) F) GDP($ billion)

```python
driver = webdriver.Chrome('chromedriver.exe')
driver.get('http://statisticstimes.com/')

time.sleep(3)
driver.find_element_by_xpath('//*[@id="top"]/div[2]/div[2]/button').click()

time.sleep(3)
driver.find_element_by_xpath('//*[@id="top"]/div[2]/div[2]/div/a[3]').click()
url = driver.current_url
time.sleep(3)
#opening the page by fetching URL (Indias stated GDP)
#using try except to deal with popups
try:
    driver.get(url)
    
    driver.find_element_by_xpath('//*[@id="top"]/div[2]/div[2]/button').click()

    time.sleep(3)
    driver.find_element_by_xpath('//*[@id="top"]/div[2]/div[2]/div/a[3]').click()

    driver.get(driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[2]/ul/li[1]/a').get_attribute('href'))
except:
    pass

#fetching all the data in table cells
rows = [row.text for row in driver.find_elements_by_xpath('//tr[@role="row"]//td')]

rank = [rank for rank in rows][0::8][:33]
state = [state for state in rows][1::8][:33]
GDP19_20 = [gdp19_20 for gdp19_20 in rows][2::8][:33]
GDP18_19 = [gdp18_19 for gdp18_19 in rows][3::8][:33]
share18_19 = [gdp18_19 for gdp18_19 in rows][4::8][:33]
GDP_Billion = [gdp_billion for gdp_billion in rows][5::8][:33]

df = pd.DataFrame({'Rank':rank, 'State':state, 'GSDP(19-20)':GDP19_20, 'GSDP(18-19)':GDP18_19, 'Share(2018)':share18_19, 'GDP($ billion)':GDP_Billion})
df
```

Scrape the details of trending repositories on Github.com. Url = https://github.com/ You have to find the following details:</br>A) Repository title B) Repository description C) Contributors count D) Language used

```python
driver = webdriver.Chrome('chromedriver.exe')
driver.get('https://github.com/')

time.sleep(3)
#clicking on explore dropdown
driver.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details').click()
#clicking on trenting
driver.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul/li[5]/a').click()
```

```python
rep_title = [title.text for title in driver.find_elements_by_xpath('//article[@class="Box-row"]//h1')]
rep_des = [desc.text for desc in driver.find_elements_by_xpath('//article[@class="Box-row"]//p')]

containers = driver.find_elements_by_xpath('//article[@class="Box-row"]')
stars = [stars.find_element_by_xpath('.//a[@class="Link--muted d-inline-block mr-3"]').text for stars in containers]

language = []
for i in containers:
    try:
        language.append(i.find_element_by_xpath('.//span[@itemprop="programmingLanguage"]').text)
    except:
        language.append('-')
```

```python
df = pd.DataFrame({'Repository title':rep_title, 'Repository description':rep_des, 'Contributors count':stars, 'Language used':language})
df
```

## Ques-6

Scrape the details of top 100 songs on billiboard.com. Url = https:/www.billboard.com/ You have to find the following details: </br>A) Song name B) Artist name C) Last week rank D) Peak rank E) Weeks on board

```python
driver = webdriver.Chrome('chromedriver.exe')
driver.get('https:/www.billboard.com/')

time.sleep(5)
#clicking on hot100 songs
driver.find_element_by_xpath('//li[@class="o-nav__list-item "]').click()

#container which contains all the information about the song
containers = driver.find_elements_by_xpath('//div[@class="o-chart-results-list-row-container"]')
time.sleep(3)

song_name = [data.find_element_by_xpath('.//h3').text for data in containers]
artist_name = [data.find_elements_by_xpath('.//li')[3].text.split('\n')[1] for data in containers]
week_on_board = [data.find_elements_by_xpath('.//li')[3].text.split('\n')[-1] for data in containers]
peak_rank = [data.find_elements_by_xpath('.//li')[3].text.split('\n')[-2] for data in containers]
last_week_rank = [data.find_elements_by_xpath('.//li')[3].text.split('\n')[-3] for data in containers]

df = pd.DataFrame({'Song name':song_name, 'Artist name':artist_name, 'Last week rank':last_week_rank, 'Peak rank':peak_rank, 'Weeks on board':week_on_board})
df
```

## Ques-7

Scrape the details of Data science recruiters from naukri.com. Url = https://www.naukri.com/ You have to find the following details: </br>A) Name B) Designation C) Company D) Skills they hire for E) Location
</br>Note: - From naukri.com homepage click on the recruiters option and the on the search pane type Data science and click on search. All this should be done through code

```python
driver = webdriver.Chrome('chromedriver.exe')
driver.get('https://www.naukri.com/hr-recruiters-consultants')

for i in range(0,2):
    driver.find_element_by_tag_name('body').send_keys(Keys.END)

time.sleep(5)
name = [data.text for data in driver.find_elements_by_xpath('//div[@class="recInfo"]//p//span')][0::3]
designation = [data.text for data in driver.find_elements_by_xpath('//div[@class="recInfo"]//p//span')][1::3]
company = [data.text for data in driver.find_elements_by_xpath('//div[@class="recInfo"]//p//a')][1::2]
skills_they_hire_for = [data.text for data in driver.find_elements_by_xpath('//div[@class="hireSec"]')]
location = [data.text for data in driver.find_elements_by_xpath('//div[@class="recInfo"]//p//span')][2::3]

df = pd.DataFrame({'Name':name, 'Designation':designation, 'Company':company, 'Skills they hire for':skills_they_hire_for, 'Location':location})
df
```

```python
driver = webdriver.Chrome('chromedriver.exe')
driver.get('https://www.imdb.com/list/ls095964455/')

#
containers = driver.find_elements_by_xpath('//div[@class="lister-item-content"]')

name = [data.find_elements_by_xpath('.//a')[0].text for data in containers]
year_span = [i.find_elements_by_xpath('.//span')[1].text for i in containers]
genre = [data.find_elements_by_xpath('.//p/span')[4].text for data in containers]
run_time = [data.find_elements_by_xpath('.//p/span')[2].text for data in containers]
ratings = [data.find_elements_by_xpath('.//div/span')[1].text for data in containers]
votes = [data.find_element_by_xpath('.//span[@name="nv"]').text for data in containers]

df = pd.DataFrame({'Name':name, 'Year span':year_span, 'Genre':genre, 'Run time':run_time, 'Ratings':ratings, 'Votes':votes})
df
```

Details of Datasets from UCI machine learning repositories. Url = https://archive.ics.uci.edu/ You have to find the following details: </br>A) Dataset name B) Data type C) Task D) Attribute type E) No of instances F) No of attribute G) Year
</br>Note: - from the home page you have to go to the ShowAllDataset page through code.

```python
driver = webdriver.Chrome('chromedriver.exe')
driver.get('https://archive.ics.uci.edu/')


time.sleep(5)
#click on view all the datasets button
driver.find_element_by_xpath('/html/body/table[1]/tbody/tr/td[2]/span[2]').click()

table = driver.find_element_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]')

#all the data in //p tags
data = table.find_elements_by_xpath('.//p[@class="normal"]')
data_text = [i.text for i in data]

dataset_name = data_text[0::7]
data_type = data_text[1::7]
task = data_text[2::7]
attribute_type = data_text[3::7]
no_of_instances = data_text[4::7]
no_of_attribute = data_text[5::7]
year = data_text[6::7]

df = pd.DataFrame({'Dataset name':dataset_name, 'Data type':data_type, 'Task':task, 'Attribute type':attribute_type, 'No of instance':no_of_instances, 'No of Attribute':no_of_attribute, 'Year':year})
df
```
