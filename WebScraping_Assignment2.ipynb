### Ques 1 Naukri.com ( Job Position - Data Analyst, Job Location - Bangalore)

```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
import time
```

```python
driver = webdriver.Chrome('chromedriver.exe')
time.sleep(2)
driver.get('https://www.naukri.com/')

#Entering Data in Fields
search_field_designation = driver.find_element_by_class_name('suggestor-input ')
search_field_designation.send_keys('Data Analyst')
search_field_locator = driver.find_element_by_xpath('/html/body/div/div[2]/div[3]/div/div/div[3]/div/div/div/input')
search_field_locator.send_keys('Bangalore')

#search button
search_button = driver.find_element_by_xpath('/html/body/div/div[2]/div[3]/div/div/div[6]')
search_button.click()
time.sleep(5)

title_tags = driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")
company_tags = driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']")
location_tags = driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']/span[1]")
experience_tags = driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi experience']/span[1]")

job_titles = [data.text for data in title_tags]
location = [data.text for data in location_tags]
company_name = [data.text for data in company_tags]
experience_required = [data.text for data in experience_tags]

print(len(job_titles))

df = pd.DataFrame({'job_titles':job_titles, "location":location, "company_name":company_name, "experience_required":experience_required})

df[0:10]
```

```python
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.keys import Keys

import time
from selenium.webdriver.support.ui import Select
```

```python
driver = webdriver.Chrome('chromedriver.exe')
driver.get('http://www.ambitionbox.com/')

time.sleep(3)

#clicking on salaries field
driver.find_element_by_xpath('//a[@class="link salaries"]').click()


time.sleep(3)
#entering data into field and clicking on data science dropdown
job_profile = driver.find_element_by_id('jobProfileSearchbox').send_keys('Data Scientist')
time.sleep(6)
driver.find_element_by_xpath('/html/body/div/div/div/main/section[1]/div[2]/div[1]/span/div/div/div[1]').click()
time.sleep(4)

#empty lists
companies = []
salaries = []
experience = []

average_salary = []
minimum_salary = []
maximum_salary = []

time.sleep(4)
#loops for appending data in the list
#spliting data by "\n" and restricting unwanted data with 
for data in driver.find_elements_by_xpath('//div[@class="company-info"]//a'):
    companies.append( data.text.split('\n')[0])
for data in driver.find_elements_by_xpath('//div[@class="company-info"]')[0:10]:
    salaries.append( data.text.split('\n')[1])
for data in driver.find_elements_by_xpath('//div[@class="company-info"]')[0:10]:
    experience.append( data.text.split('\n')[-1])

#loops for salary data
#spliting data by "\n" and restricting unwanted data with 
for data in driver.find_elements_by_xpath('//div[@class="salary-range-wrapper"]'):
    average_salary.append(data.text.split('\n')[0])
for data in driver.find_elements_by_xpath('//div[@class="salary-range-wrapper"]'):
    minimum_salary.append(data.text.split('\n')[1]) 
for data in driver.find_elements_by_xpath('//div[@class="salary-range-wrapper"]'):
    maximum_salary.append(data.text.split('\n')[-1]) 


print(len(average_salary),len(minimum_salary),len(maximum_salary))

df = pd.DataFrame({'company_name':companies, 'total salary record':salaries, 'average_salary':average_salary, 'minimum_salary':minimum_salary, 'maximum_salary':maximum_salary, 'experience_required':experience})
df
```
