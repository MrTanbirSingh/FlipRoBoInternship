### Ques-2

2. In the above question, now scrape the following details of each product listed in first 3 pages of your
search results and save it in a data frame and csv. In case if any product has less than 3 pages in search
results then scrape all the products available under that product name. </br>
Details to be scraped are: "BrandName", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" 
and "Product URL". In case, if any of the details are missing for any of the product then replace it by "-".

```python
#using input function to get input from user
keyword = input('Please Enter keword you want to search: ')
driver = webdriver.Chrome('chromedriver.exe')
driver.get('http://www.amazon.in/')

#passing the input in the search field entered by the user
search_field_mainpage = driver.find_element_by_id("twotabsearchtextbox").send_keys(keyword)

time.sleep(3)
#clicking on search button
button_mainpage =  driver.find_element_by_id('nav-search-submit-button').click()

```

```python
brand_name = []
name = []
price = []
expected_delivery = []
availability = []
product_url = []

container = driver.find_elements(By.XPATH,'.//div[@class="a-section a-spacing-base"]')

for next_page in range(1,4):
    
    #fetching titles = Name
    for data in container:
        try:
            name.append(data.find_element_by_xpath('.//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]//span').text)
        except:
            name.append('-')
    time.sleep(3)

    
    # brand names
    for data in container:
        try:
            brand_name.append(data.find_element_by_xpath('.//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]//span').text.split()[0])
        except:
            brand_name.append('-')
 

    #fetching prices from the container
    for i in driver.find_elements(By.XPATH,'//div[@class="a-section a-spacing-base"]'):
        try:
            price.append(i.find_element_by_xpath('.//span[@class="a-price-whole"]').text)
        except:
            price.append('-')
    
    
    time.sleep(3)
    #Date of Arrival
    for data in container:
        try:
            expected_delivery.append(data.find_element_by_xpath('.//div[@class="a-row s-align-children-center"]').text)
        except:
            expected_delivery.append('-')
    #Availability

    time.sleep(3)
    #Fetching Product URL
    for data in container:
        try:
            product_url.append(data.find_element_by_xpath('.//a[@class = "a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]').get_attribute('href'))
        except:
            product_url.append('-')           
    
    time.sleep(6)
    #doing to the next page
    driver.get('https://www.amazon.in/s?k=Guitar&page='+str(next_page))

#checking if product is in stock. If the price is not available for the product it means that the product is not available.
for p in price:
    if p != '-':
        availability.append('In Stock')
    else:
        availability.append('-')    
```

```python
#Saving the data in .csv file.

df = pd.DataFrame({'brand_name':brand_name, 'name':name, 'price':price, 'expected_delivery':expected_delivery, 'availability':availability, 'product_url':product_url})
df.to_csv('Guitar_data_amazon.csv')
df.sample(5)
```

7. Write a program to scrap all the available details of best gaming laptops from digit.in.

```python
delay = 5
driver = webdriver.Chrome('chromedriver.exe')
driver.get("https://www.digit.in/")

click_laptops = driver.find_element_by_xpath('/html/body/div[1]/div/div[4]/ul/li[3]/a').click()

WebDriverWait(driver, delay).until(EC.presence_of_element_located((By.XPATH, '/html/body/div[6]/div/div[2]/div[2]/ul/li[10]/a')))
open_best_gaming = driver.get(driver.find_element_by_xpath('/html/body/div[6]/div/div[2]/div[2]/ul/li[10]/a').get_attribute('href'))

product_name = []
info = {}
price = []

for i in range(1,11):
    product_name.append(driver.find_element_by_xpath(f'//div[@data-index="{i}"]//h3').text)
    price.append(driver.find_element_by_xpath(f'//table[@id="summtable"]//tr[{i}]//td[3]').text)
    
column_names = re.split('[:\n]',driver.find_elements_by_class_name('Spcs-details')[1].text)[1::2]

for i in re.split('[:\n]',driver.find_elements_by_class_name('Spcs-details')[1].text)[1::2]:
    info[i.rstrip()] = []
    
for count in range(0,len(column_names)):
    for i in driver.find_elements_by_class_name('Spcs-details'):
        info[list(info.keys())[count]].append((re.split('[:\n]',i.text)[1:][1::2][count]))
        
df = pd.DataFrame(info)
df.insert(0,'name',product_name)
df.insert(1,'price',price)
df.to_csv('Top_gaming_laptops.csv')
df
```

### Ques-8

Write a python program to scrape the details for all billionaires from www.forbes.com. Details to be scrapped: “Rank”, “Name”, “Net worth”, “Age”, “Citizenship”, “Source”, “Industry”.


```python
driver = webdriver.Chrome('chromedriver.exe')
driver.get('http://www.forbes.com/')

billionaires_link = driver.find_element_by_xpath('/html/body/div[1]/header/nav/div[3]/ul/li[1]/div[2]/ul/li[2]/a').get_attribute('href')
driver.get(billionaires_link)

time.sleep(10)

try:
    #close the popup
    driver.find_element_by_xpath('//button[@class="tp-close tp-active"]').click()
except:
    pass


Rank = [i.text.replace('.','') for i in driver.find_elements_by_xpath('//div[@class="rank"]')]
Name = [i.text.replace('.','') for i in driver.find_elements_by_xpath('//div[@class="personName"]')]
NetWorth = [i.text.replace('.','') for i in driver.find_elements_by_xpath('//div[@class="netWorth"]')]
Age = [i.text.replace('.','') for i in driver.find_elements_by_xpath('//div[@class="age"]')]
Citizenship = [i.text.replace('.','') for i in driver.find_elements_by_xpath('//div[@class="countryOfCitizenship"]')]
Source = [i.text.replace('.','') for i in driver.find_elements_by_xpath('//div[@class="source"]')]
Industry = [i.text.replace('.','') for i in driver.find_elements_by_xpath('//div[@class="netWorth"]')]

df= pd.DataFrame({'Rank':Rank, 'Name':Name, 'NetWorth':NetWorth, 'Age':Age, 'Citizenship':Citizenship, 'Source':Source, 'Industry':Industry})
df.to_csv('Billionairs.csv')
df
```

```python
driver = webdriver.Chrome('chromedriver.exe')
driver.get('http://www.hostelworld.com/')

driver.find_element_by_xpath('//input[@id="location-text-input-field"]').click()
time.sleep(3)
driver.find_element_by_xpath('//input[@class="search-input"]').send_keys('London')
time.sleep(4)
driver.find_element_by_xpath('//*[@id="predicted-search-results"]/li[2]/div').click()
driver.find_element_by_xpath('//button[@id="search-button"]').click()

time.sleep(6)

hostel_name = [i.text for i in driver.find_elements_by_xpath('//h2[@class="title title-6"]')]
distance = [i.text for i in driver.find_elements_by_xpath('//span[@class="description"]')]
private_price = [i.text.split('\n')[1].replace('Rs','') for i in driver.find_elements_by_xpath('//a[@class="prices"]')]
dorm_price = [i.text.split('\n')[-1] for i in driver.find_elements_by_xpath('//a[@class="prices"]')]
ratings = [i.text.split('\n')[0] for i in driver.find_elements_by_xpath('//div[@class="bottom-rating"]')]
total_ratings = [i.text.split('\n')[-1] for i in driver.find_elements_by_xpath('//div[@class="bottom-rating"]')]

description = []
description = driver.find_elements_by_xpath('//div[@class="property"]')
description_values = []

for i in description:
    if len(i.find_element_by_xpath('.//div[@class="info"]').text.split('\n')) > 5:
        description_values.append(i.find_element_by_xpath('.//div[@class="info"]').text.split('\n')[:3])
    else:
        description_values.append('-')
        
df = pd.DataFrame({'HostelName':hostel_name, 'distance':distance, 'description':description_values, 'private_price':private_price, 'dorm_price':dorm_price, 'ratings':ratings, 'total_ratings':total_ratings})
df
```
